```python
#Bar plot for SHOOTERS and X position
plt.figure(figsize=(10,6)) 
shots = sns.barplot(x='SHOOTER',y='X', data=nba)

#Russell Westbrook shoots better in the X position than the other 3 players
```

```python
#Scatter plot for SHOOTERS and X position
plt.figure(figsize=(10,6))
shots = sns.scatterplot(x='SHOOTER', y='X', data=nba, size='X')
```

```python
#Bar plot for SHOOTERS and Y Position
plt.figure(figsize=(10,6))
shots = sns.barplot(x='SHOOTER', y='Y', data=nba)

#Seth Curry shoots better in the Y position than the other 3 players
```

```python
#Scatter plot for SHOOTERS and Y Position
plt.figure(figsize=(10,6))
shots = sns.scatterplot(x='SHOOTER', y='Y', data=nba, size='Y')
```

```python
#We want to explore at what range each shooters shoot best

#X position
plt.figure(figsize=(10,6))
r_comparison = sns.barplot(data=nba, x='SHOOTER', y='X', hue='RANGE')

#We see that Seth Curry shoots best from the '(15,19)' range, Chris Paul from
#'(20,24)' range, Russell Westbrook from '(20,24)' range, Trae Young from
#'(5,9)' range. All are from the X position.
```

```python
#Y position
plt.figure(figsize=(10,6))
r_comparison = sns.barplot(data=nba, x='SHOOTER', y='Y', hue='RANGE')

#Looking at the graph, we see that all players are best in the range, '(25,29)'
#from the Y position. However, we see that Russell Westbrook has a higher
#shooting range in '(10,14)' than the other three players.
```

```python
#Looking at the players total scores made and missed

#X position
plt.figure(figsize=(10,6))
a_score = sns.catplot(data=nba, x='SHOOTER', y='X', hue='SCORE', kind='bar' )
```

```python
#Y position
plt.figure(figsize=(10,6))
a_score = sns.catplot(data=nba, x='SHOOTER', y='Y', hue='SCORE', kind='bar')
```

```python
#Let's try with the volin plot
plt.figure(figsize=(10,6))
a_score = sns.violinplot(x=nba.SHOOTER, y=nba['X'])
plt.show()
```
